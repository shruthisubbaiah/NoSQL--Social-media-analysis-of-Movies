# What does our data look like?

The data displayed above has a lot of dependencies and repitations, also it is not clean. We would further break the data into tables and clean the data a bit to proceede with our opperations.

```python
#main table - All the other tables will be connected to this one
movies = pd.DataFrame()
movies[['movie_id','movie_names','movie_description','imdb_ratings','metascores','runtime','gross_value','year_release']] = movie_data[['Movie_id','movie_names','movie_description','imdb_ratings','metscores','runtime','gross_value','year_release']]
movies = movies.drop_duplicates()
movies = movies.reset_index(drop=True)
movies.head()
```

```python
#Actors main table - this will connect to movie table using a seperate relational table
actors = pd.DataFrame()
actors[['movie_names','starcast_name']] = movie_data[['movie_names','star_cast']]
actors = actors.drop_duplicates()
actors = actors.reset_index(drop = True)
actors.head()
```

```python
#director main table - this will connect to movie table using a seperate relational table
directors = pd.DataFrame()
directors[['movie_names','director_name']] = movie_data[['movie_names','director_name']]
directors = directors.drop_duplicates()
directors = directors.reset_index(drop = True)
directors.head()
```

```python
#genre main table - this will connect to movie table using a seperate relational table
genres = pd.DataFrame()
genres[['movie_names','genre']] = movie_data[['movie_names','genre']]
genres = genres.drop_duplicates()
genres = genres.reset_index(drop = True)
genres.head()
```

```python
#posts main table - this will connect to movie table using a seperate relational table (these are posts from facebook)
posts = pd.DataFrame()
posts[['movie_id','post_id','created_at','post_message','likes_count','share_count','comments_count','user_engagement']] = post_data[['movie_id','post_id','created_at','post_message','post_likes_count','post_shares_count','post_comment_count','user_engagement']]
posts = posts.drop_duplicates()
posts['created_at'] = posts['created_at'].astype('datetime64[ns]')
posts = posts.reset_index(drop = True)
posts.head()
```

```python
#comments main table - this will connect to posts table using a seperate relational table
comments = pd.DataFrame()
comments[['post_id','comments_id','created_at','comments']] = comment_data[['post_id','comments_id','created_at','post_comments']]
comments = comments.drop_duplicates()
comments['created_at'] = comments['created_at'].astype('datetime64[ns]')
comments = comments.reset_index(drop = True)
comments.head()
```

```python
#tweets main table - this will connect the tweets to the movies
tweets = pd.DataFrame()
tweets[['movie_id','tweet_id','tweet_text', 'created_date', 'retweet_count']] = tweets_data[['movie','tweet_id','tweet_text','created_date','retweet_count']]
tweets = tweets.drop_duplicates()
tweets = tweets.reset_index(drop = True)
tweets.head()
```

```python
tag_scores = tag_scores.drop_duplicates()
tag_scores = tag_scores.reset_index(drop = True)
tag_scores.head()
```

```python
#Filling all the NA variables with 0 to avoid any sort of conflicts while calling the program.
posts["likes_count"] = posts['likes_count'].fillna(0).astype(int)
posts["comments_count"] = posts['comments_count'].fillna(0).astype(int)
posts["share_count"] = posts['share_count'].fillna(0).astype(int)
posts["user_engagement"] = posts['user_engagement'].fillna(0).astype(int)
tweets["retweet_count"]=tweets['retweet_count'].fillna(0).astype(int)
```

# Design Explanation:

If it were a SQL database we could have stored the values into the database using the relational tables and replicate the orignal database using JOIN queries.

Since we want to create a NoSQL database, we want to store the data in form of documments. Before we dig depper into how to create them, let us take a look at what documents and collections actully mean in terms of database designs.

## Collections:

Collections in NoSQL are analogus to the Tables in the SQL databases. Insert, Update and Delete operations can be performed on a collection.

## Documents:

Each record of a master table in Relational database can be considered as a document here. For example in our movie database we have data for each movie and we store all the other tables in relation with the key givven in the movies tables. In the NoSQL database, each movie will be a stand alone doccument on its own.

## Creation of the documents:

The below code is used for converting the tables into documents which will further be stored in a collection.
The below structure could be understood as a list of nested dictionaries to store data related to a movie.



```python
#Defining empty dictionaries and list which will hold the data inside.
movie_doc = []
movie_dict = {}
actor = []
director = []
genre = []
posts_dict ={}
comments_dict = {}
comments_list = []
posts_list = []
tweets_dict = {}
tweets_list = []
tags = []
tags_dict = {}

#Creating a single document for a movie.
for i in range(len(movies)):
    movie_dict = {"movie_name":movies["movie_names"][i],
                 "description": movies["movie_description"][i],
                 "ratings": float(movies["imdb_ratings"][i]),
                 "metascores": float(movies["metascores"][i]),
                 "runtime" : float(movies["runtime"][i]),
                 "gross_value": float(movies["gross_value"][i]),
                 "year_release": float(movies["year_release"][i])}
    # the values being directly stored under the movie names are unique values to each movie.
    
    #For the given movie, we search for the actors and store that information within the list of actors
    for j in range(len(actors)):
        if actors["movie_names"][j] == movies["movie_names"][i]:
            act = {"actor" : actors["starcast_name"][j]}
            actor.append(act)
            
    actors_list = {"actors":actor}
    movie_dict.update(actors_list) 
    actor = []
    
    #Here we store the names of the directors for the movie.
    for k in range(len(directors)):
        if directors["movie_names"][k] ==movies["movie_names"][i]:
            director = directors["director_name"][k]
    
    direct = {"director":director}
    movie_dict.update(direct)
    director = []
    
    #Here we are storing the genres associated with a movie.
    for l in range(len(genres)):
        if genres["movie_names"][l] == movies["movie_names"][i]:
            g = {"gener" :genres["genre"][l]}
            genre.append(g)
            
    gen = {"genre":genre}
    movie_dict.update(gen)
    genre = []
    
    #Here we are storing all the posts related to a movie page.
    for m in range(len(posts)):
        if posts["movie_id"][m] == movies["movie_id"][i]:
            posts_dict = {"created_at": posts["created_at"][m],
                        "post_message": posts["post_message"][m],
                        "likes": int(posts["likes_count"][m]),
                        "shares": int(posts["share_count"][m]),
                        "comments_count": int(posts["comments_count"][m]),
                        "user_engagement": int(posts["user_engagement"][m])}
            
            #We further loop the comments related to a particular post.
            for n in range(len(comments)):
                if comments["post_id"][n] == posts["post_id"][m]:
                    comments_dict = {"created_at": comments["created_at"][n],
                                   "comment": comments["comments"][n]}
                    comments_list.append(comments_dict)
                    
        cmnt = {"comments":comments_list}
        posts_dict.update(cmnt)
        posts_list.append(posts_dict)
        comments_list = []
                
    pst = {"posts": posts_list}
    movie_dict.update(pst)
    posts_list = []
    
    
    #Here we store the tweets for a particular movie.
    for o in range(len(tweets)):
        if tweets["movie_id"][o] == movies["movie_id"][i]:
            tweets_dict = {"tweet": tweets["tweet_text"][o],
                          "created_date": tweets["created_date"][o],
                          "retweet": int(tweets["retweet_count"][o])}
            tweets_list.append(tweets_dict)     
    
    twt = {"tweets": tweets_list}
    movie_dict.update(twt)
    tweets_list = []
    
    #here we store the tags and scores for the tags associated with a particular movie.
    for p in range(len(tag_scores)):
        if tag_scores["movie_id"][p] == movies["movie_id"][i]:
            tags_dict = {"word": tag_scores["word"][p],
                        "score":float(tag_scores["score"][p])}
            tags.append(tags_dict)    
    
    tag = {"tags": tags}
    movie_dict.update(tag)
    tags = []
    
    #Here we are saving all the documemts into a list of documments also known as the collection.
    movie_doc.append(movie_dict)
```

In the above code, we define a dictonary "movie_dict" this is the main dictionary which will become a document in our NoSQL database. We create one such dict for each movie which is further appended in a list named "movie_doc". This list of movie_dicts will become our collection in the NoSQL database.

# Design decisions:

The same way we have many to many relationships and many to one relationships in the relational databases, we construct a similar concept in the document database.

We can embeed all the data in one dictionary, or create various collections and reference one collection to the other.

##### Golden Rule
**_"Things that are queried together and operated on together should be stored together."_**

This is one of the most important concepts of designing databases in NoSQL.

## Use Case:

As we previously saw, in our database we have used the embedding as opposed to refferencing. Now, before we get into why we did that, let us try and understand why we reference the documents at the first place.

Embeding of the data is usually done on data with same voaltility, being queried together and have the same kind of relationships. How embeding helps is, for a query to work the code has to make just one round trip to get all the data related to the document.

Let us consider an example of Fcaebook, the details such as User names, User details may not change over time that much. But, the things like comments, posts, likes, shares etc. may change much more frequently. If we have all the posts embeded for a particular user in the same tab, everytime one has to change something related to the post will have to query the user each time. Imagnine doing that for a million users, too computationaly expensive right?. 

Thus, for such cases we use referencing where we have a seperate document for posts and we just reference the post_id to the user information document.

# 2. How viral are my posts?

We take a weighted sum of the count of likes, shares and comments on my posts. The maximum weightage is given to the share count as that signifies the max spreading of any post, followed by the likes and comments.

```python
i=0
viral_movie=pd.DataFrame(columns=['movie_name','Virality'])
for m in db.Movies.find({}):
    total_likes=0
    total_shares=0
    total_comments=0
    for p in m.get("posts"):
        total_likes=total_likes+p.get("likes")
        total_shares=total_shares+p.get("shares")
        total_comments += p.get("comments_count")
    viral_movie.loc[i]=[m.get("movie_name"),total_likes*(0.3)+total_comments*(0.2)+total_shares*(0.5)]
    i=i+1
viral_movie.sort_values(by = "Virality",ascending = False).head()
```
